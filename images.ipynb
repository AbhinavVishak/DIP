```python
import cPickle as pickle
import numpy as np


import matplotlib.pyplot as plt
%matplotlib inline

from project_nn import LogisticRegression
from project_nets import plot_logistic
from project_utils import *

datasets = load_mnist(theano_shared=False)

test_set_x, test_set_y = datasets[2]


f = open('best_logreg.pkl','rb')
classifier = pickle.load(f)
f.close() 

b = classifier.b.get_value()
w = classifier.W.get_value()

print('Ground Truth')
print( test_set_y[0:10] )

print('Predictions')
print(np.argmax(np.dot( test_set_x[0:10,:] , w ) + b , axis = 1 ) )
    
mis = np.reshape(test_set_x[8] , (28,28 ) )
cor = np.reshape(test_set_x[9] , (28,28 ) )

plt.subplot(1,2,1)
imgplot = plt.imshow( mis )
plt.title('Wrong( GT-5 , pred-6)')
imgplot.set_cmap('Greys')

plt.subplot(1,2,2)
imgplot2 = plt.imshow( cor )
plt.title('Correct - 9 ')
imgplot2.set_cmap('Greys')

plt.savefig('logreg_image.png')
```
